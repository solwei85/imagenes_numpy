```python
#Ejercicio 3
array3= np.array([[2, 3, 4, 5, 6], [2, 2, 5, 10, 1], [2, 13, 24, 55, 6], [12, 30, 44, 5, 16], [12, 23, 34, 45, 66]])
print(array3)
print (array3.max())
```

```python
print (array3.max(axis=0)) #lista con los maximos por cada columna. Ej: columna 1: 12, columna 2: 30, etc
print (array3.max(axis=1)) #lista con los maximos por cada fila
```

```python
#Ejercicio 4 Broadcasting
import random
array4= np.random.random_sample((2,2))
print(array4)
```

```python
print(array4+1) #1 no tiene la misma dimension que array4, pero por broadcasting lo podemos hacer. 
#En el fondo esta sumando a una matrix de 2x2 con valores de 1.
print(array4*2) #tambien se puede hacer operaciones como multiplicacion
```

```python
array4 + np.array([2,3])
#tambien funciona con otro array de diferente dim.
```

```python
#Ejercicio 5 Index
print(array4 [0,0] + array4[1,1])
#suma del primer valor de la primera lista + 2do valor de la 2da lista.
```
